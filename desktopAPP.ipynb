```python
import PySimpleGUI as sg

#appの色を指定
sg.theme("DarkBrown3")

#翻訳したい言語を設定
selects = ["英語","フランス語","ドイツ語"]

#appのレイアウトを指定（T:ラベル、ML：マルチライン（複数行）のテキストボックス、Im：画像、Combo、コンボボックス、B：ボタン）
layout = [[sg.T("入力文:"),
          sg.ML("こんにちは。私はGeminiの勉強をしています",s=(50,3),k="in")],
          [sg.Im("icon_100100_16.png"),
          sg.Combo(selects,default_value=selects[0],s=10,k='cb'),
          sg.T("に翻訳するよ"),
          sg.B("実行",k="btn")],
          [sg.ML(k="txt",font=(None,14),s=(60,13))]]

#appの名前とサイズ、フォントの種類、サイズを指定
win = sg.Window("アプリテスト",layout,font=(None,14),size=(550,400))

#実行するための関数を作る
def execute():
    prompt = f"以下の文書を{v['cb']}に翻訳してください。\n### {v['in']}"
    win["txt"].update(prompt)

#関数を無限ループにする（常に動いている状態、ボタンを押せば起動するのは無限ループにさせないと。通常のアプリが起動したら動かせるのはそういうこと）
#e:イベント、
while True:
    e,v = win.read()
    if e == "btn":
        execute()
    if e == None:
        break

#windowsを閉じる(イベントがなかったら、　if文からブレイクし、クローズする）
win.close()
```

```python
import PySimpleGUI as sg
import os
import google.generativeai as genai

#APIKEY、使用するモデルを設定
GOOGL_API_KEY = os.environ['GOOGL_API_KEY']
genai.configure(api_key=GOOGL_API_KEY)
model = genai.GenerativeModel('gemini-pro')

#appの色を指定
sg.theme("DarkBrown3")

#翻訳したい言語を設定
selects = ["Python","JavaScript","Go","pine","MQL4"]

#appのレイアウトを指定（T:ラベル、ML：マルチライン（複数行）のテキストボックス、Im：画像、Combo、コンボボックス、B：ボタン）
layout = [[sg.T("処理:"),
          sg.ML("１から１０までの合計を求める。",s=(50,3),k="in")],
          [sg.Im("icon_100100_16.png"),
          sg.Combo(selects,default_value=selects[0],s=10,k='cb'),
          sg.T("のプログラムを書くよ！"),
          sg.B("実行",k="btn")],
          [sg.ML(k="txt",font=(None,14),s=(60,13))]]

#appの名前とサイズ、フォントの種類、サイズを指定
win = sg.Window("プログラム自動生成",layout,font=(None,14),size=(550,400))

#実行するための関数を作る
def execute():
    prompt = f"以下の処理を{v['cb']}のプログラムを書いてください。\n### {v['in']}"
    #モデルの出力結果を受け取る
    response = model.generate_content(prompt)
    #前の文字をクリア
    win["txt"].update("")
    #レスポンスのテキストで反映（文字数等いろんなものが返ってくるので、答え（テキスト）だけを文字で反映させる
    win["txt"].update(response.parts[0].text)
    
#関数を無限ループにする（常に動いている状態、ボタンを押せば起動するのは無限ループにさせないと。通常のアプリが起動したら動かせるのはそういうこと）
#e:イベント、
while True:
    e,v = win.read()
    if e == "btn":
        execute()
    if e == None:
        break

#windowsを閉じる(イベントがなかったら、　if文からブレイクし、クローズする）
win.close()
```

```python
import PySimpleGUI as sg
import os
import google.generativeai as genai

#APIKEY、使用するモデルを設定
GOOGL_API_KEY = os.environ['GOOGL_API_KEY']
genai.configure(api_key=GOOGL_API_KEY)
model = genai.GenerativeModel('gemini-pro')

with open('業務要件.txt', 'r')as f:
    fdata = f.read()
    

#appの色を指定
sg.theme("DarkBrown3")

#翻訳したい言語を設定
selects = ["嗜好調査","栄養教室","推定栄養量","食品検索","料理検索"]

#appのレイアウトを指定（T:ラベル、ML：マルチライン（複数行）のテキストボックス、Im：画像、Combo、コンボボックス、B：ボタン）
layout = [[sg.T("業務命令:"),
          sg.ML("ここに入力してください。",s=(50,3),k="in")],
          [sg.Im("icon_100100_16.png"),
          sg.Combo(selects,default_value=selects[0],s=10,k='cb'),
          sg.T("でアイディアを出すよ"),
          sg.B("実行",k="btn")],
          [sg.ML(k="txt",font=(None,14),s=(60,13))]]

#appの名前とサイズ、フォントの種類、サイズを指定
win = sg.Window("AI栄養士さん",layout,font=(None,14),size=(550,400))

#実行するための関数を作る
def execute():
    role = "あなたは病院勤務の管理栄養士です。\n"
    prompt = f"{v['cb']}の業務に関しての{v['in']}の質問に対して、以下の条件に従って答えてください。\n "
    prompt = role + prompt + fdata
    #モデルの出力結果を受け取る
    response = model.generate_content(prompt)
    #前の文字をクリア
    win["txt"].update("")
    #レスポンスのテキストで反映（文字数等いろんなものが返ってくるので、答え（テキスト）だけを文字で反映させる
    win["txt"].update(response.parts[0].text)
    
#関数を無限ループにする（常に動いている状態、ボタンを押せば起動するのは無限ループにさせないと。通常のアプリが起動したら動かせるのはそういうこと）
#e:イベント、
while True:
    e,v = win.read()
    if e == "btn":
        execute()
    if e == None:
        break

#windowsを閉じる(イベントがなかったら、　if文からブレイクし、クローズする）
win.close()
```
